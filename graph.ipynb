```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
df = pd.read_csv("C:/Users/shrof/OneDrive/Documents/Documents/UVA/Summer 2024 Semester/Programming/MSDS-Orientation-Computer-Survey.csv")
```

```python
ram = df['RAM (in GB)']
ram = np.array(ram)
```

```python
ram = ram[ram < 129]
```

```python
ram
```

```python
x_axis = np.arange(0,65,1)

fig = plt.figure(figsize=(8,6),facecolor='white')
ax = fig.add_subplot()

plt.hist(ram, label = 'Ram (Gb)', bins = 10)

ax.set_xlabel('RAM (GB)', fontsize=16)
ax.set_ylabel('Count', fontsize=16)
ax.set_title('Histogram of Ram (GB) for 2021-2024 MSDS Cohort',fontsize=18)
plt.yticks(fontsize=14)
plt.xticks(fontsize = 14)

plt.show()
```
